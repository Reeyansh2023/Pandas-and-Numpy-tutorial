```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

# import data

```python
df=pd.read_csv("gun-violence-data.csv")
df1=df.copy()
df1.head()
```

```python
df1.info()
```

```python
df1=df1.iloc[:,:7]
df1.head(10)
```

```python
df1.drop(columns=["incident_id","city_or_county","address"],inplace=True)
```

```python
df1.head()
```

```python
df1['day']=pd.to_datetime(df1["date"]).dt.day
```

```python
df1['month']=pd.to_datetime(df1["date"]).dt.month
```

```python
df1['year']=pd.to_datetime(df1["date"]).dt.year
```

```python
df1["quarter"]=pd.to_datetime(df1["date"]).dt.quarter
```

```python
df1["day_of_week_num"]=pd.to_datetime(df1["date"]).dt.dayofweek
```

```python
df1["day_of_week_name"]=pd.to_datetime(df1["date"]).dt.day_name()
```

#### 2nd method day of week name
######  replace numeric day of week value with text


```python
df1['day_of_week_num']=df1['day_of_week_num'].map({0:'Monday',1:'Tuesday',2:'Wednesday',3:'Thrusday',4:'Friday',5:'Saturday',6:'Sunday'})
```

```python
df1['month']=df1['month'].map({1:'January',2:'Febuary',3:'March',4:'April',5:'May',6:'June',7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'})
```

```python
df1.head()
```

```python
df1.drop('date',axis=True,inplace=True)
```

```python
df1['total_victim']=df1['n_killed']+df1['n_injured']#Sum up number of deaths and injuries
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.groupby('year')['total_victim'].sum() # value_counts,mean() ,catorical/numerical numerical
```

```python
df1.groupby('month')['total_victim'].count()
```

```python
df1.groupby('day_of_week_num')['total_victim'].count()
```

```python
# what are the total number of victims from gun violence for each US state?
# we first group the number of victims by stae then sum them all up
df1.groupby('state')['total_victim'].count()[4:8].sort_values(ascending=False)
```
